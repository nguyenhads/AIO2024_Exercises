# Visualize Activation Funciton

## Importing

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

## Visualization

```python
# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))


def relu(x):
    return np.maximum(0, x)


def elu(x, alpha=0.5):
    return np.where(x >0, x, alpha * (np.exp(x) - 1))

    
def plot(x, y, activation_name="sigmoid", figsize=(8,6)):

    _, ax = plt.subplots()
    ax.plot(x, y, "r")
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    ax.set_title(f"{activation_name.capitalize()}")
    
    plt.show()
```

```python
# sigmoid
x_arr = np.arange(-20, 20)
out = sigmoid(x_arr)

plot(x=x_arr, y=out, activation_name="sigmoid")
```

```python
# relu
x_arr = np.arange(-20, 20)
out = relu(x_arr)

plot(x=x_arr, y=out, activation_name="relu")
```

```python
# elu
x_arr = np.arange(-10, 10)
out = elu(x_arr)

plot(x=x_arr, y=out, activation_name="elu")
```
